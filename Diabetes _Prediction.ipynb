```python
import pandas as pd
```

```python
data = pd.read_csv("diabetes_prediction_dataset.csv")
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns",None)
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data.apply(lambda x: x.str.strip() if x.dtype == 'object' else x)
```

```python
data['gender'].unique()
```

```python
data['gender'] = data['gender'].str.replace("Other","Female")
```

```python
data['gender'].value_counts()
```

```python
data['heart_disease'].unique()
```

```python
chn = {
    1: "high blood Pressure",
    0: "Normal blood pressure"
}
data['hypertension']=data['hypertension'].replace(chn).astype("object")
data.sample(3)
```

```python
data['heart_disease'].unique()
```

```python
data['heart_disease']=data['heart_disease'].map({1:'Heart Disease',0:'No Heart Disease'}).astype("object")
data.sample(3)
```

```python
data['smoking_history'].unique()
```

```python
data['smoking_history']=data['smoking_history'].map({'never':'Never','No Info':'No Information','current':'Currently','former':'Former','ever':'Ever','not current':"Not Currently"})
data.sample(3)
```

```python
data = data.rename(columns={'bmi':'BMI_Level'})
data.columns
```

```python
data['HbA1c_level'].unique()
```

```python
data['blood_glucose_level'].unique()
```

```python
data.info()
```

```python
X=data.drop(columns=['diabetes'])
y=data['diabetes']
X=X.dropna()
y=y.dropna()
X.shape, y.shape
```

```python
print(X.isnull().sum())
print(y.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X['age'] = imputer.fit_transform(X[['age']])
X['BMI_Level'] = imputer.fit_transform(X[['BMI_Level']])
X['HbA1c_level'] = imputer.fit_transform(X[['HbA1c_level']])
```

```python
X.sample(3), y.sample(3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y) 
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import make_column_transformer
```

```python
cat_col = ['gender','hypertension','heart_disease','smoking_history']
num_col = ['age','BMI_Level','HbA1c_level','blood_glucose_level']
```

```python
make_col = make_column_transformer((OneHotEncoder(handle_unknown="ignore",categories='auto'),cat_col),
                                (StandardScaler(with_mean=False),num_col),
                                remainder="passthrough"
)
make_col
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
regression = LogisticRegression()
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(make_col, regression)
pipe
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy : ", accuracy)
```
